```python
from matplotlib.patches import Ellipse
import matplotlib.pyplot as plt
%matplotlib inline
```

<h3>Se requiere crear una clase Elipse, la cual el constructor sea con valores iniciales cualesquiera. Adicionalmente, la clase debe tener los siguientes métodos:</h3>

```python
class Elipse(object):

    def __init__(self, width=2, height=3, color='r', border = 'y'):
        self.height = height
        self.width = width
        self.color = color
        self.border = border

    def widthSetter(self, w):
        self.width = w

    def heightSetter(self, h):
        self.height = h

    def colorSetter(self, c):
        self.color = c

    def borderSetter(self, b):
        self.border = b

    def drawEllipse(self):
        plt.gca().add_patch(Ellipse((0, 0), 20, 15, facecolor = self.color, edgecolor = self.border))
        plt.axis('scaled')
        plt.show()


                    


    
```

```python
Elipse1 = Elipse(5,10,'red', 'blue')
Elipse1.drawEllipse()
```
